```python
def adder(a, b = 0, *abhijeet, **kwargs):
    print(b)
    a = a + b
    for i in abhijeet:
        a += int(i)
    print(kwargs)
    print(abhijeet)
    for i in kwargs:
        print(i, kwargs[i])
    return a
```

```python
def adder(a, b):
    return a + b
```

```python
def login_requ(abcd):
    
    def wrapper(username, password, *args):
        if username in users and users[username] == password:
            print('Authenticated')
            print(abcd(*args))
        else:
            print(username, password)
            print('Try again')
    
    return wrapper 
```

```python
prot_adder = login_requ(adder)
```

```python
prot_adder('Sarthak', 'password', 1, 5)
```

```python
adder = login_requ(adder)
```

```python
adder('Abhijeet', '12345678', 50, 60)
```
