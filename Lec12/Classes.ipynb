```python
class Data_Science:
    
    subject = 'Data_Science'     # Attributes
    mentor = 'Abhijeet'
    place = 'MM school'
    students = []
    
    def temp(self, abcd):                          # Methods
        print('Data_Science')
        print('Hello')
        print(abcd)
        print(self.place)
        
    def update_place(self, new_place):
        self.place = new_place
        
    def update_student(self, new_stu):
        self.students.append(new_stu)
```

```python
# Data_Science.temp(876453, 45631)
```

```python
ds1 = Data_Science()
ds2 = Data_Science()
```

```python
ds1.temp("Abhijeet")
```

```python
ds1.update_place('Coding Blocks')
```

```python
print(id(ds1.place), id(ds2.place), id(Data_Science.place), id(Data_Science().place))
print(ds1.place, ds2.place, Data_Science.place, Data_Science().place)
```

```python
Data_Science.students.append('Sarthak')
Data_Science.students.append('Pratham')
```

```python
ds1.update_student('Harsh')
```
